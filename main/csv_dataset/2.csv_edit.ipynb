```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
images = pd.read_csv('./정형데이터/train/images_train.csv', )
annotations = pd.read_csv('./정형데이터/train/annotations_train.csv')
categories = pd.read_csv('./정형데이터/train/categories_train.csv')
```

```python
images
```

```python
pattern = r'(.+?)_(.+?)\.jpg'
images[['Vienna_Code','etc']] = images['fileName'].str.extract(pattern)
images = images.drop(columns=['etc'])
images
```

```python
images.rename(columns={'id':'image_id'}, inplace=True)
images
```

```python
images.info()
```

```python
images.to_csv('./images_trian_v1.csv', index=False)
```

```python
from collections import OrderedDict

# 이미지 데이터프레임에서 'Vienna Code' 열의 고유한 값 추출하고 순서를 유지
unique_values = list(OrderedDict.fromkeys(images['Vienna_Code']))

# 결과 출력
print(unique_values)
```

```python
categories['Vienna_Code'] = unique_values
categories
```

```python
categories.rename(columns={'id': 'category_id'}, inplace=True)
categories
```

```python
categories.to_csv('./categories_train_v1.csv', index=False)
```
