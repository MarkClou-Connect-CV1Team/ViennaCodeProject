```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
images = pd.read_csv('./정형데이터/train/images_train_v1.csv', )
annotations = pd.read_csv('./정형데이터/train/annotations_train_v1.csv')
categories = pd.read_csv('./정형데이터/train/categories_train_v1.csv')
```

```python
images
```

```python
categories
```

```python
merge1 = pd.merge(images, categories, on='Vienna_Code', how = 'inner')
merge1
```

```python
merge2 = pd.merge(annotations, merge1, on=['image_id','name'], how='inner')
merge2
```

```python
temp = pd.DataFrame(merge2['Vienna Code'].drop_duplicates())
temp['Code_Index'] = temp.reset_index().index
temp
```

```python
merge3 = pd.merge(merge2, temp, on='Vienna Code', how = 'inner')
merge3
```

```python
merge3.to_csv('./train.csv', index=False)
```
