```python
SEED = 42
BATCH_SIZE = 8
EPOCHS = 5
IMGSZ = 960
MODEL = "v"
```

```python
# model = YOLO(f"{MODEL}/train/weights/last.pt")
model = YOLO("yolov8n.yaml")

results = model.train(
    data="/content/drive/MyDrive/data/yolocustom.yaml",
    imgsz=IMGSZ,
    epochs=EPOCHS,
    batch=BATCH_SIZE,
    patience=5,
    workers=16,
    device=0,
    exist_ok=True,
    project=f"{MODEL}",
    name="train",
    seed=SEED,
    pretrained=False,
    resume= True,
    optimizer="Adam",
    lr0=1e-3,
    augment=True,
    val=True,
    cache=True
    )
```
